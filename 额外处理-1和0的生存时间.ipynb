```python
import pandas as pd
```

```python
# 读取C:\MyProject\MIMIC-IPE\加了中介变量的数据集\用于COX_sur的特征数据集.csv
data = pd.read_csv(r'C:\MyProject\MIMIC-IPE\加了中介变量的数据集\用于COX_sur的特征数据集.csv')
# 读取C:\MyProject\MIMIC-IPE\加了中介变量的数据集\用于统计的完整数据集.csv
data2 = pd.read_csv(r'C:\MyProject\MIMIC-IPE\加了中介变量的数据集\用于统计的完整数据集.csv')
# 读取C:\MyProject\MIMIC-IPE\加了中介变量的数据集\用于特征选择的数据集.csv
data3 = pd.read_csv(r'C:\MyProject\MIMIC-IPE\加了中介变量的数据集\用于特征选择的数据集.csv')
```

```python
# 按照Subject_id合并data和data2，，只保留data2中的Admittime，Deathtime和Dod列
data = pd.merge(data, data2[['Subject_id', 'Admittime', 'Deathtime', 'Dod']], on='Subject_id')

# 按照
```

```python
# 对于data3中D180D列不为0的行，使用Deathtime列的时间减去Admittime列的时间，以天做单位，精确到小数点后1位，替换T.O.D_180d列的值，如果Deathtime列的值为NaN，则使用Dod列的值.Apply this operation only to rows where 'D180D' is not 0。T.O.D_180d列中空值填充为180，大于180的值更改为180
# Convert to datetime if not already
data3['Admittime'] = pd.to_datetime(data3['Admittime'])
data3['Deathtime'] = pd.to_datetime(data3['Deathtime'])
data3['Dod'] = pd.to_datetime(data3['Dod'])

# Calculate the difference in days and round to 1 decimal place
data3.loc[data3['D180D'] != 0, 'T.O.D_180d'] = (data3['Deathtime'] - data3['Admittime']).dt.total_seconds() / (24 * 60 * 60)
data3['T.O.D_180d'] = data3['T.O.D_180d'].round(4)

# Replace 'T.O.D_180d' with 'Dod' where 'Deathtime' is NaN
data3.loc[data3['Deathtime'].isna() & (data3['D180D'] != 0), 'T.O.D_180d'] = (data3['Dod'] - data3['Admittime']).dt.total_seconds() / (24 * 60 * 60)
data3['T.O.D_180d'] = data3['T.O.D_180d'].round(1)


# Fill NaN values with 180 and replace values greater than 180 with 180
data3['T.O.D_180d'].fillna(180, inplace=True)
data3.loc[data3['T.O.D_180d'] > 180, 'T.O.D_180d'] = 180
```

```python
# Convert to datetime if not already
data3['Admittime'] = pd.to_datetime(data3['Admittime'])
data3['Deathtime'] = pd.to_datetime(data3['Deathtime'])
data3['Dod'] = pd.to_datetime(data3['Dod'])

# List of target columns and their corresponding limits
target_cols = [('T.O.D_90d', 90), ('T.O.D_60d', 60), ('T.O.D_31d', 31)]

for col, limit in target_cols:
    # Calculate the difference in days and round to 1 decimal place
    data3.loc[data3['D180D'] != 0, col] = (data3['Deathtime'] - data3['Admittime']).dt.total_seconds() / (24 * 60 * 60)
    data3[col] = data3[col].round(1)

    # Replace 'col' with 'Dod' where 'Deathtime' is NaN
    data3.loc[data3['Deathtime'].isna() & (data3['D180D'] != 0), col] = (data3['Dod'] - data3['Admittime']).dt.total_seconds() / (24 * 60 * 60)
    data3[col] = data3[col].round(1)

    # Fill NaN values with limit and replace values greater than limit with limit
    data3[col].fillna(limit, inplace=True)
    data3.loc[data3[col] > limit, col] = limit
```

```python
# data3保存到C:\MyProject\MIMIC-IPE\加了中介变量的数据集\用于特征选择的数据集.csv
data3.to_csv(r'C:\MyProject\MIMIC-IPE\加了中介变量的数据集\用于特征选择的数据集.csv', index=False)
```

```python
# 对于D180D列不为0的行，使用Deathtime列的时间减去Admittime列的时间，以天做单位，精确到小数点后1位，替换T_O_D_180d列的值，如果Deathtime列的值为NaN，则使用Dod列的值.Apply this operation only to rows where 'D180D' is not 0。T_O_D_180d列中空值填充为180，大于180的值更改为180
# Convert to datetime if not already
data['Admittime'] = pd.to_datetime(data['Admittime'])
data['Deathtime'] = pd.to_datetime(data['Deathtime'])
data['Dod'] = pd.to_datetime(data['Dod'])

# Calculate the difference in days and round to 1 decimal place
data['T_O_D_180d'] = (data['Deathtime'] - data['Admittime']).dt.total_seconds() / (24 * 60 * 60)
data['T_O_D_180d'] = data['T_O_D_180d'].round(4)

# Replace 'T_O_D_180d' with 'Dod' where 'Deathtime' is NaN
data.loc[data['Deathtime'].isna(), 'T_O_D_180d'] = (data['Dod'] - data['Admittime']).dt.total_seconds() / (24 * 60 * 60)
data['T_O_D_180d'] = data['T_O_D_180d'].round(1)

# Apply this operation only to rows where 'D180D' is not 0
data.loc[data['D180D'] != 0, 'T_O_D_180d'] = data['T_O_D_180d']

# T_O_D_180d列中空值填充为180，大于180的值更改为180
data['T_O_D_180d'].fillna(180, inplace=True)
data.loc[data['T_O_D_180d'] > 180, 'T_O_D_180d'] = 180
```

```python
# 删除data中Admittime，Deathtime和Dod列
data.drop(columns=['Admittime', 'Deathtime', 'Dod'], inplace=True)
# 保存data到C:\MyProject\MIMIC-IPE\加了中介变量的数据集\用于COX_sur的特征数据集.csv
data.to_csv(r'C:\MyProject\MIMIC-IPE\加了中介变量的数据集\用于COX_sur的特征数据集.csv', index=False)
```
