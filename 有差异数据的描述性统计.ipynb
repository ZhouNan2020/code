```python
import pandas as pd
```

```python
# 读取C:\MyProject\MIMIC-IPE\加了中介变量的数据集\用于统计的完整数据集.csv
data = pd.read_csv(r'C:\MyProject\MIMIC-IPE\加了中介变量的数据集\用于统计的完整数据集.csv')
'''
分组变量是Group，
以下是描述性统计中呈现出差异的变量：
Age（连续变量）
Blood_related_diseases（无序多分类变量）
DTI（二分类）
FXaI（二分类）
UFH（二分类）
VKA（二分类）
Tot.ICU_hrs（连续变量）
Cnt.ICU（连续变量）
LOHS_hrs（连续变量）
F.Hosp（二分类）
'''
# 以分组变量Group为分组，对以上变量使用合适的图进行展示，要求图形风格统一使用ggplot，图的内容要丰富

import seaborn as sns
import matplotlib.pyplot as plt

# Set the ggplot style with ticks for a more academic look
sns.set_style("ticks")

# Set the color palette to blue and red
palette = sns.color_palette(["#3498db", "#e74c3c"])

# List of variables
variables = ['Age', 'Blood_related_diseases', 'DTI', 'FXaI', 'UFH', 'VKA', 'Tot.ICU_hrs', 'Cnt.ICU', 'LOHS_hrs', 'F.Hosp']

# For each variable
for var in variables:
    # If the variable is continuous
    if data[var].dtype in ['int64', 'float64']:
        plt.figure(figsize=(10, 6))
        sns.boxplot(x='Group', y=var, data=data, palette=palette)
        plt.title(f'Boxplot of {var} grouped by Group')
        plt.show()
    # If the variable is categorical
    else:
        plt.figure(figsize=(10, 6))
        sns.countplot(x=var, hue='Group', data=data, palette=palette)
        plt.title(f'Countplot of {var} grouped by Group')
        plt.show()
```
